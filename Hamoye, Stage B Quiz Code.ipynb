# MACHINE LEARNING REGRESSION

```python
# Drop the columns: date and lights before further analysis  
new_df = df.drop(columns = ['date', 'lights'])
```

```python
# normalize the dataset by importing MinMaxScaler
# then check the first 5 rows of the normalized dataframe
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalized_df = pd.DataFrame(scaler.fit_transform(new_df), columns = new_df.columns)
normalized_df.head()
```

# Question 12

```python
# From the dataset, fit a linear model on the relationship between the temperature in the living room
# in Celsius (x = T2) and the temperature outside the building (y = T6).
```

```python
# Check the feature_df 
feature_df = normalized_df.drop(columns = ['T6'])
feature_df.head()
```

```python
# check the target variable
target_var = normalized_df['T6']
target_var.head()
```

```python
#  train_test_split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(feature_df, target_var , test_size = 0.3, random_state = 42)
```

```python
# Linear Regression
from sklearn.linear_model import LinearRegression
lin_mod = LinearRegression()
lin_mod.fit(x_train, y_train) 
```

```python
# Predictions
pred_val = lin_mod.predict(x_test)
pred_val
```

```python
# R-Squared 
from sklearn.metrics import r2_score
r2 =  round(r2_score(y_test, pred_val), 2)
print('The R^2 value in two decimal places is ', r2)
```

# Question 13

```python
# Mean Absolute Error
from sklearn.metrics import mean_absolute_error
mae= round(mean_absolute_error(y_test, pred_val), 2)
print('The Mean Absolute Error in two decimal places is ',mae)
```

# Question 14

```python
# Residual Sum of Squares(RSS)
import numpy as np
rss = round(np.sum(np.square(y_test - pred_val)), 2)
print('The Residual Sum of Squares in two decimal places is ', rss)
```

# Question 15

```python
# Root Mean Square Error (RMSE)
from sklearn.metrics import mean_squared_error
rmse = round(np.sqrt( mean_squared_error(y_test, pred_val)), 3)
print('The Root Mean Squared Error in three decimal places is ', rmse)
```

# Question 16

```python
# R-Squared 
from sklearn.metrics import r2_score
r2 =  round(r2_score(y_test, pred_val), 2)
print('The Coefficient of determination in two decimal places is ', r2)
print('The Coefficient of determination is also known as the R-Square')
```

```python
# Ridge Regression
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha = 0.4)
ridge_reg.fit(x_train, y_train)
rsme = round(np.sqrt( mean_squared_error(y_test, pred_val)), 3)
print('Root Mean Square Error:', rsme)
r2 =  r2_score(y_test, pred_val)
print('R-square :', r2)
print()
print('With this result, the Root Mean Square Error did not change.')
```

# Question 20

```python
# Feature Selection and Lasso Regression
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha = 0.001)
lasso_reg.fit(x_train, y_train)
rsme = round(np.sqrt( mean_squared_error(y_test, pred_val)), 3)
print('Root Mean Square Error:', rsme)
r2 =  r2_score(y_test, pred_val)
print('R-square :', r2)
print('the new Root Mean Square Error with Lasso Regression is ',  rsme)
```

```python
# The feature weights are obtained from the linear model above
```

```python
# comparing the effects of regularization
from sklearn.linear_model import LinearRegression
def get_weights_df(model, feat, col_name):
    # this function returns the weight of every feature 
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df

lin_mod_weights = get_weights_df(lin_mod, x_train, 'Lin_Mod_Weight')
ridge_weights_df = get_weights_df(ridge_reg, x_train, 'Ridge_Weight')
lasso_weights_df = get_weights_df(lasso_reg, x_train, 'Lasso_Weight')

final_weights = pd.merge(lin_mod_weights, ridge_weights_df, on = 'Features')
final_weights = pd.merge(final_weights, lasso_weights_df, on = 'Features')
final_weights
```
